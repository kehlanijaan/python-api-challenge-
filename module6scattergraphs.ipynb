```python
import matplotlib.pyplot as plt
Latitude = [18.3851, 58.0948, -42.7826, 42.097, 36.0564,-54.8, 42.6001, -24.8667, 66.9395, -21.3393, -22.8375, 59.7035, -3.2778, 58.4619, 77.484]
Temperature = [22.06, 13.3, 11.72, 5.77, 14.53, 8.81, 10.36, 20.02, 2.72, 22.82, 20.36, 8.94, 31.07, 5.63, -17.32]
plt.scatter (Latitude, Temperature)
plt.xlabel("Latitude")
plt.ylabel("Temperature")
plt.show()
import numpy as np
#obtain m(slope) and b(intercept) of linear regression line
m, b = np.polyfit(Latitude, Temperature, 1)
#add linear regression line to scatter plot
plt.plot(Latitude, m*(Latitude+b))
plt.xlabel("Latitude")
plt.ylabel("Temperature")
```
